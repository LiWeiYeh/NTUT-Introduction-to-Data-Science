## Imports

```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

## Main Program

```python
# reads the dataframe
df = pd.read_csv('advertisement.csv', na_values = ':')

# splits df into train and test sets
train, test = train_test_split(df, test_size=0.2)

df
```

```python
df.mean()
```

```python
# object for the class
model = LinearRegression()
```

### (Cost, Click)

```python
x = df[['Cost']]
y = df[['Click']]

# perform linear regression
model.fit(x, y)
y_hat = model.predict(x)
plt.plot(x, y,'o', alpha = 0.5)
plt.plot(x, y_hat, 'r', alpha = 0.5)
plt.xlabel('cost')
plt.ylabel('click')
print("y=aX+b => a coefficient:", model.coef_)
print("y=aX+b => b coefficient:", model.intercept_)
print("R square", metrics.r2_score(y_hat, y))
```

```python
# prediction of the amount of clicks for 25000 cost

# 25000 cost
X = [[25000]]
y_hat = model.predict(X)
print("Prediction of clicks for 25000 cost:", y_hat)
```
