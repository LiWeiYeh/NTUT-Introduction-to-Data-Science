# Step 6
### Plot the distribution of three types of Abalone using different colors where x-axis is the diameter and y-axis the whole weight

```python
Male_Data = df[df['sex'] == 'M'] 
Female_Data = df[df['sex'] == 'F'] 
Infant_Data = df[df['sex'] == 'I']

# kwargs = dict(alpha=0.5, bins=100)

# x1 = df.loc[df.sex=='M', 'diam']
# x2 = df.loc[df.sex=='F', 'diam']
# x3 = df.loc[df.sex=='I', 'diam']

# #y1 = df.loc[df.sex=='M', 'whole']
# #y2 = df.loc[df.sex=='F', 'whole']
# #y3 = df.loc[df.sex=='I', 'whole']

# plt.hist(x1, **kwargs, color='g', label='M')
# plt.hist(x2, **kwargs, color='b', label='F')
# plt.hist(x3, **kwargs, color='r', label='I')
# plt.gca().set(title="Distribution of Abalone's diameter by sex", 
#               xlabel='diameter')

# plt.legend();

Male_Diam = Male_Data['diam']
Male_Whole_Weight = Male_Data['whole']

Female_Diam = Female_Data['diam']
Female_Whole_Weight = Female_Data['whole']

Infant_Diam = Infant_Data['diam']
Infant_Whole_Weight = Infant_Data['whole']

plt.scatter(Male_Diam, Male_Whole_Weight, color='red')
plt.scatter(Female_Diam, Female_Whole_Weight, color='green')
plt.scatter(Infant_Diam, Infant_Whole_Weight, color='blue')
plt.xlabel('Diameter')
plt.ylabel('Whole Weight')
plt.show()
```

# Step 7
### Show the proportion of three types of Abalone using pie chart.

```python
labels = 'Male', 'Female', 'Infant'

Amount_Male = len(Male_Data.index)
Amount_Female = len(Female_Data.index)
Amount_Infant = len(Infant_Data.index)

Amount_Data = Amount_Male + Amount_Female + Amount_Infant

fracs = [Amount_Male / Amount_Data, 
         Amount_Female / Amount_Data, 
         Amount_Infant / Amount_Data]

plt.pie(x=fracs, labels=labels, autopct='%1.2f%%', startangle=90)
plt.show
```
