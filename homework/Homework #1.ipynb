### 1) Calculate the sum of the multiples of 2 or 3 between the numbers 1 and N, including 1 and N

```python
# For loop
def SumMultiplesOf2Or3ForLoop(N):
    sum = 0
    for i in range (1, N + 1):
        # if divisible by 2, 3 OR if it's 1 OR if it's N
        if( ((i % 2) == 0) or (i % 3) == 0 or (i == 1) or (i == N)):
            sum += i
    return sum

print(SumMultiplesOf2Or3ForLoop(25))

# While loop
def SumMultiplesOf2Or3WhileLoop(N):
    sum = 0
    counter = 1
    while counter <= N:
        if( ((counter % 2) == 0) or (counter % 3) == 0 or (counter == 1) or (counter == N)):
            sum += counter
            
        counter += 1
    return sum

print(SumMultiplesOf2Or3WhileLoop(25))

# # Recursive Function
# def Rec_SumMultipleOf2Or3(N):
#     if N <= 0:
#         return 0
#     else:
#         if( ((N % 2) == 0) or (N % 3) == 0 or (N == 1)):
#             return N + Rec_SumMultipleOf2Or3(N - 1)
#         else:
#             return Rec_SumMultipleOf2Or3(N - 1)
# print(Rec_SumMultipleOf2Or3(25))
```

### 2) Print out the multiplication table

```python
def MultiplicationTableForLoop(n):
    
    # Too big number messes up the formatting
    if n > 31:
        print("Please pick a number the same as or smaller than 31")
        
    else:
        # finds the string length of the largest number in the set
        stringLengthOfNSquared = len(str(n * n))
        
        # loops through to print the rows
        for i in range (1, n + 1):
            # initializes the row
            row = ""
            # loops through to add the numbers to the row with added spaces for formatting
            for j in range (1, n + 1):
                row += (str(i*j) + ((stringLengthOfNSquared - len(str(i*j))) * " ") + " ")
            print(row)
            
MultiplicationTableForLoop(31)
```

### 3) Check if N is a prime number or not

```python
def isPrimeWhileLoop(N):

    if N <= 0:
        print("Pick number greater than 0")
    elif N == 1:
        print("%d is NOT a prime number" % N)
    else:
        counter = 2
        isPrime = True
        while counter < N:
            # if N is divisible by the previous number excluding 0 and 1 (counter starts at 2)
            if N % counter == 0:
                # isPrime is equal to false
                isPrime = False
                break
            counter += 1
        if isPrime == True:
            print("%d is a prime number" % N)
        else:
            print("%d is NOT a prime number" % N)    
    
# for i in range(0, 1000):
#     isPrimeForLoop(i);
isPrimeWhileLoop(23)
isPrimeWhileLoop(39)
```

### 4) Check if N is a Narcissistic number ot not

```python
def isNarcissistic(N):
    stringN = str(N)
    # finds the length of the string
    lengthN = len(str(N))
    
    sum = 0
    
    # loops through the string length of N
    for i in range (0, lengthN):
        # add the i-th number of (N to the power of N)
        sum += int(stringN[i])**lengthN
    
    # if the sum is equal to N, it's a Narcissistic number
    if sum == N:
        print("%d is a Narcissistic number" % N)
    else :
        print("%d is NOT a Narcissistic number" % N)
    
for i in range(0, 1000):
    isNarcissistic(i)
```
